```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# 현재 파이썬 스크립트 위치 기준으로 '500 cyc' 폴더 경로 설정
base_folder_path = os.path.join(os.getcwd(), '500 cyc')

# Define the folders (subfolders within '500 cyc')
folders = ['1-No-Xray', '2-2min', '3-1hour']
colors = ['red', 'blue', 'green']  # Colors for each folder

# Function to filter data
def filter_data(df):
    filtered_df = df[df['Cycle Index'] > -1]
    filtered_df.reset_index(drop=True, inplace=True)
    return filtered_df

# Initialize a figure
fig, ax = plt.subplots(figsize=(fig_width, fig_height))

for folder, color in zip(folders, colors):
    # 폴더 경로 설정
    folder_path = os.path.join(base_folder_path, folder)
    
    # List all Excel files in the folder
    files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]
    files_sort = sorted(files)
    
    # Initialize a DataFrame to collect all data
    all_data = pd.DataFrame()
    
    for file in files_sort:
        file_path = os.path.join(folder_path, file)
        # Read the Excel file
        df = pd.read_excel(file_path, header=0, sheet_name='cycle')
        filtered_df = filter_data(df)
        
        all_data = pd.concat([all_data, filtered_df], ignore_index=True)
    
    # Calculate the mean DChg. Cap.(Ah) for each Cycle Index
    mean_data = all_data.groupby('Cycle Index')['DChg. Cap.(Ah)'].mean().reset_index()
    
    # Plot the mean data
    plt.plot(mean_data['Cycle Index'], mean_data['DChg. Cap.(Ah)'], color=color)

# Adjust plot settings
ax.set_xlabel('Cycle number')  # X축 타이틀
ax.set_ylabel('Average Discharge Capacity [Ah]')  # Y축 타이틀

ax.grid()
ax.plot([], [], linestyle='-', linewidth=2, color='red', label='No X-ray', alpha=0.5)
ax.plot([], [], linestyle='-', linewidth=2, color='blue', label='2 minutes', alpha=0.5)
ax.plot([], [], linestyle='-', linewidth=2, color='green', label='1 hour', alpha=0.5)
ax.legend()  # 레전드 추가



# 플롯을 표시
plt.show()
```

```python
# Initialize a figure
fig, ax = plt.subplots(figsize=(fig_width, fig_height))

labels = ['No X-ray', '2 minutes', '1 hour']

for k1, (folder, color) in enumerate(zip(folders, colors)):
    # 폴더 경로 설정
    folder_path = os.path.join(base_folder_path, folder)
    
    # List all Excel files in the folder
    files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]
    files_sort = sorted(files)
    
    for k2, file in enumerate(files_sort):
        file_path = os.path.join(folder_path, file)
        # Read the Excel file
        df = pd.read_excel(file_path, header=0, sheet_name='cycle')
        filtered_df = filter_data(df)
        
        # 개별 파일 데이터를 플롯 (각 샘플을 개별적으로)
        if k2 == 0:
            plt.plot(filtered_df['Cycle Index'], filtered_df['DChg. Cap.(Ah)'], color=color, label=labels[k1])
        else:
            plt.plot(filtered_df['Cycle Index'], filtered_df['DChg. Cap.(Ah)'], color=color)

# Adjust plot settings
ax.set_xlabel('Cycle number')  # X축 타이틀
ax.set_ylabel('Discharge capacity [Ah]')  # Y축 타이틀

ax.grid()
ax.legend(facecolor="white", edgecolor="white", framealpha=1)  # 레전드 추가
ax.set_xlim([0, 301])

plt.savefig("fig2.pdf", dpi=300, bbox_inches="tight", format="pdf")
# 플롯을 표시
plt.show()
```
